```python
import matplotlib.pyplot as plt
```

```python
#Basic Subplot (1 row, 2 columns)
# Data
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]
```

```python
# Create a figure with 1 row and 2 columns
plt.subplot(1, 2, 1)  # (rows, cols, plot_no)
plt.plot(x, y1)
plt.title('Double of x')
plt.grid(True)
```
