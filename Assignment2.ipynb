```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
years = [2003, 2007, 2011, 2015, 2019]
sachin_runs = [1141, 1425, 1894, 482, 0]
kohli_runs = [0, 281, 1381, 1295, 1377]
sehwag_runs = [1195, 1420, 1399, 0, 0]
width = 0.25
```

```python
x = np.arange(len(years))
plt.bar(x - width, sachin_runs, width=width, label="Sachin")
plt.bar(x, kohli_runs, width=width, label="Kohli")
plt.bar(x + width, sehwag_runs, width=width, label="Sehwag")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
plt.show()
```

```python
players = ["Sachin","Kohli","Sehwag"]
runs_5yrs = [1141+1425+1894+482+0, 0+281+1381++1295++1377, 1195+1420+1399+0+0]
```

```python
plt.barh(players,runs_5yrs,color="green")
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
plt.show()
```
